Dapat dikatakan bahwa listrik telah menjadi sumber energi utama dalam setiap kegiatan baik di

&nbsp; &nbsp; &nbsp; &nbsp;Berdasarkan kondisi tersebut, maka pada penelitian ini akan dirancang suatu sistem
simulasi untuk mengontrol AC pada ruangan-ruangan dengan
menggunakan konsep fuzzy logic.</p></font>

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl

#import digunakan untuk menggunakan fungsi-fungsi yang berada dalam modul, 
#pernyataan import terdiri dari sebuah kata kunci dan nama sebuah modul.

#numpy = (modul untuk komputas ilmiah)
#skfuzzy = (modul yang digunakan untuk logika fuzzy)
#matplotlib.pyplot = (modul yang digunakan untuk menyajikan data dalam bentuk grafik/visual)
#control = (modul yang berada dalam skfuzzy)
```

<body><font size="3">
   <p>&nbsp; &nbsp; &nbsp; &nbsp;Variabel input fuzzy untuk mengontrol lampu dan AC adalah berdasarkan intensitas
cahaya, banyak orang, temperatur udara dan banyak lampu. Masing-masing variabel terdiri dari beberapa
       himpunan fuzzy, yaitu:</p>
    <div class="alert alert-success" role="alert">
       <ol type='a'>
           <li>Intensitas cahaya : Gelap, Normal, Terang</li>
           <li>Banyak Orang      : Sedikit, Banyak</li>
           <li>Temperatur Udara  : Dingin, Normal, Panas</li>
           <li>Banyak Lampu : Sedikit, Banyak</li>
       </ol></div>
   <p>&nbsp; &nbsp; &nbsp; &nbsp;Sementara untuk variabel output fuzzy terdiri adalah Temperatur AC.
      Yang terdiri dari beberapa himpunan fuzzy, yaitu:</p>
       <div class="alert alert-warning" role="alert">
       <ol type='a'>
           <li>Temperatur AC : Dingin, Sejuk, Panas</li>
           </ol></div>
    </font>
</body>

```python
IC = ctrl.Antecedent (np.arange(1,4000,1), 'Intensitas Cahaya')
BO = ctrl.Antecedent (np.arange(1,30,1), 'Banyak Orang')
TU = ctrl.Antecedent (np.arange(1,45,1), 'Temperatur Udara')
BLM = ctrl.Antecedent (np.arange(1,10,1), 'Banyak Lampu Menyala')
TA = ctrl.Consequent(np.arange(1,34,1), 'Temperatur AC')
```

<font size="3"><p>Disini terdapat 4 Input dan 1 Output</p><br/>
&nbsp; &nbsp; &nbsp; &nbsp;Variabel yang memiliki berfungsi sebagai input dapat kita lihat dengan syntax <i><b>ctrl.Antecedent</b></i> dan variabel output <i><b>ctrl.Consequent</b></i>, setiap variabel memiliki rentang nilai yang berbeda yang dapat dilihat pada syntax <i><b>np.arange</b></i>, pada syntax tersebut terdapat 3 bagian nilai yang dapat diartikan dengan <font color="green">start</font>,<font color="red">stop</font>,<font color="blue">jump</font>.
<p>&nbsp; &nbsp; &nbsp; &nbsp;Sebagai contoh variabel IC dimulai dari 1 dan nilai akhirnya adalah 4000 dari nilai ke 1 ke nilai berikutnya itu dilalui dengan 1 lompatan nilai, begitu juga dengan variabel yang lainnya.</p></font>

<h2>Himpunan Fuzzy</h2>

<font size="3"><p>&nbsp; &nbsp; &nbsp; &nbsp;&nbsp; &nbsp;Fungsi keanggotaan adalah grafik yang mewakili besar dari derajat keanggotaan masing-masing variabel input yang berada dalam interval antara 0 dan 1. Derajat keanggotaan sebuah variabel x dilambangkan dengan simbol μ(x). Rule- rule menggunakan fungsi keanggotaan sebagai faktor bobot untuk menentukan pengaruhnya pada saat melakukan inferensi untuk menarik kesimpulan.
<br/><br/>
1. Fungsi keanggotaan kurva linear<br/>
&nbsp; &nbsp; &nbsp; &nbsp;Pada grafik keanggotaan linear, sebuah variabel input dipetakan ke derajat keanggotaannya dengan digambarkan sebagai suatu garis lurus.</p>
<p>Ada 2 grafik keanggotaan linear, yaitu:<br/>
    <b>a. Grafik keanggotaan kurva linear naik</b><br/>&nbsp; &nbsp; &nbsp; &nbsp;Kenaikan himpunan dimulai pada nilai domain yang memiliki derajat keanggotaan nol bergerak ke kanan menuju ke nilai domain yang memiliki derajat keanggotaan lebih tinggi</font>

```python
nilai = [0, 100]
titik = ['a','b']

x = range(len(nilai))
plt.xticks(x, titik)
plt.plot(nilai)
plt.title('CONTOH BENTUK GRAFIK KURVA LINEAR NAIK')
plt.show()
```

<h3><div class="alert alert-info" role="alert">
Linear Naik
$μ[X] = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq a \\
       {(x-a)\over(b-a)} & \mbox{; } a ≤ x ≤ b \\
       1  &\mbox{; } x \geq b
    \end{array}
\right.$ 
</div>
</h3>

<font size="3"><b>b. Grafik keanggotaan kurva linear turun</b><br/>&nbsp; &nbsp; &nbsp;Garis lurus dimulai dari nilai domain dengan derajat keanggotaan tertinggi pada sisi kiri kemudian bergerak menurun kiri, ke nilai domain yang memiliki
derajat keanggotaan lebih rendah</font>

```python
nilai = [100, 0]
titik = ['a','b']

x = range(len(nilai))
plt.xticks(x, titik)
plt.plot(nilai)
plt.title('CONTOH BENTUK GRAFIK KURVA LINEAR TURUN')
plt.show()
```

<h3><div class="alert alert-success" role="alert">
Linear Turun
$μ[X] = 
\left\{
    \begin{array}{ll}
       (b-x)\over(b-a) & \mbox{; } a \leq x \leq b \\
       0 & \mbox{; } x \geq b
    \end{array}
\right.$
</div><h/3>

<font size="3"><b>2. Fungsi keanggotaan kurva segitiga</b><br/>&nbsp; &nbsp; &nbsp;Grafik keanggotaan kurva segitiga pada dasarnya merupakan gabungan antara 2 garis</font>

```python
nilai = [0,50,0]
titik = ['a','b','c']

x = range(len(nilai))
plt.xticks(x, titik)
plt.plot(nilai)
plt.title('CONTOH BENTUK GRAFIK KURVA SEGITIGA')
plt.show()
```

<h3><div class="alert alert-warning" role="alert">
Kurva Segitiga
$μ[X] = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq a OR  x \geq c \\
       {(x-a)\over(b-a)} & \mbox{; } a ≤ x ≤ b \\
       {(c-x)\over(c-b)} &  \mbox{; } b \leq x \leq c
    \end{array}
\right.$ 
</div></h3>

<p><font size="3"><b>3. Fungsi keanggotaan kurva trapesium</b><br/>
&nbsp; &nbsp; &nbsp;Grafik kenaggotaan kurva trapesium pada dasarnya seperti bentuk segitiga, hanya saja ada
    beberapa titik yang memiliki nilai keanggotaan 1</font></p>

```python
nilai = [0,50,50,0]
titik = ['a','b','c','d']

x = range(len(nilai))
plt.xticks(x, titik)
plt.plot(nilai)
plt.title('CONTOH BENTUK GRAFIK KURVA TRAPESIUM')
plt.show()
```

<h3>
<div class="alert alert-danger" role="alert">
Kurva Trapesium
$μ[X] = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq a, OR,  x \geq c \\
       {(x-a)\over(b-a)} & \mbox{; } a ≤ x ≤ b \\
       {1} & \mbox{; } b ≤ x ≤ c \\
       {(d-x)\over(d-c)} &  \mbox{; } x \geq d
    \end{array}
\right.$ 
</div>
</h3>

<h2><center>Persamaan Intensitas Cahaya</center></h2>
<div class="alert alert-success" role="alert">
1. $IC[Gelap] =
\left\{
        \begin{array}{12}
            {500-x \over 500-200} & \mbox { ; }  200 \leq x \leq 500 \\
            0 & \mbox { ; }  x \geq 500
        \end{array}
\right.$
</div>

<div class="alert alert-danger" role="alert">
2. $IC[Normal] = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 300, OR,  x \geq 1000 \\
       {(x-300)\over(500-300)} & \mbox{; } 300 ≤ x ≤ 500 \\
       {1} & \mbox{; } 500 ≤ x ≤ 1000 \\
       {(1500-x)\over(1500-1000)} &  \mbox{; } x \geq 1500
    \end{array}
\right.$ 
</div>

<div class="alert alert-info" role="alert">
3.  $IC[Terang] = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 1500 \\
       {(x-1500)\over(3000-1500)} & \mbox{; } 1500 ≤ x ≤ 3000 \\
       1  &\mbox{; } x \geq 3000
    \end{array}
\right.$ 
</div>

```python
IC ['Gelap'] = fuzz.trapmf(IC.universe,[1,1,200,500])
IC ['Normal'] = fuzz.trapmf(IC.universe,[300,500,1000,1500])
IC ['Terang'] = fuzz.trapmf(IC.universe,[1500,3000,4000,4000])
IC.view()
```

<font size="3">Pada variabel <b>IC</b> atau <b>Intensitas Cahaya</b> terdapat 3 himpunan fuzzy yaitu : <font color="blue">Gelap</font>,<font color="orange">Normal</font>,<font color="green">Terang</font>.<br/>
<p>maksud dari syntax <b><i>fuzz.trapmf</i></b> adalah untuk menggambarkan grafik trapesium dengan nilai tertentu yang dapat dilihat pada syntax <b><i>IC.universe</i></b> yang terdapat 4 nilai karena trapesium memiliki 4 titik sudut.</p>
<p><font color="blue" >- Variabel Gelap memiliki rentang nilai dari 0-500</font></p>
<p><font color="orange" >- Variabel Normal memiliki rentang nilai dari 300-1500</font></p>
<p><font color="green" >- Variabel Terang memiliki rentang nilai dari 1500-4000</font></p>
<p><font > Terdapat 4 nilai pada sebuah grafik dengan syntax <b><i>fuzz.trapmf</i></b> yang berarti tiap sudut (trapesium) memiliki sebuah nilai </font></p></font>
