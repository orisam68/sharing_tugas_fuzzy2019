# Penentuan Kelayakan Penerbitan Paper Menggunakan Fuzzy Logic Metode Mamdani (Studi Kasus Universitas Malikussaleh)

```python
#memanggil library yang akan digunakan
import numpy as np#Numpy Array yang akan digunakan untuk menyimpan nilai variabel yang saya inisiasi menjadi "np"
import skfuzzy as fuzz#Skfuzzy adalah module untuk Fuzzy Logic yang saya inisiasi menjadi "fuzz".import library fuzzy gunakan fuzz untuk menggunkaan library ini
from skfuzzy import control as ctrl#skfuzzy adalah library fuzzy logic pada python yang bekerja dengan numpy array
```

Faktor-faktor yang perlu diperhatikan dalam menentukan kelayakan penerbitan paper, antara lain:

1. Publish #kelayakan untuk dapat diterbitkan paper adalah telah mempublikasikan banyak paper
2. Isi & Bahan #kelayakan untuk dapat diterbitkan paper adalah isi & bahan yang bagus sehingga pembacanya merasa ingin        mengetahui hasil penelitian tersebut 
3. Penyampaian #kelayakan untuk dapat diterbitkan paper adalah tat bahasa penyampaian yang mudah dimengerti oleh pembaca      sehingga menarik untuk membacanya
4. Citation #kelayakan untuk dapat diterbitkan paper adalah nilai acuan untuk mendukung penelitian selanjutnya diperoleh besar

Guna fuzzy dalam kasus ini adalah sebagai decision making dalam menentukan kelayakan penerbitan paper menggunakan metode mamdani dengan menggunakan 2 fungsi yaitu trapeziodal dan triangular.

# Variabel Input (Antecedent)

1.Publish

$$Sering = 2-10$$

2.Isi & Bahan
$$Tidak Bagus = 0-3$$
$$Cukup = 1-6$$

4.Citation

$$Sedang = 2-6$$

# Variabel Output (Consequent)

$$Tidak Layak = 0-2$$

$$Layak = 4-10$$

# Fuzzyfikasi

Merupakan proses untuk mengubah input sistem yang mempunyai nilai tegas menjadi variabel linguistik menggunakan fungsi keanggotaan yang disimpan dalam basis pengetahuan fuzzy

```python
PB = ctrl.Antecedent(np.arange(0, 11, 1), 'publish')#variabel publish dg nilai 0 samapai 10 dengan range 1,Antecedent adalah library untuk 'Input'
IB = ctrl.Antecedent(np.arange(0, 11, 1), 'isi & bahan')#variabel isi & bahan dg nilai 0 samapai 10 dengan range 1,Antecedent adalah library untuk 'Input'
PP = ctrl.Antecedent(np.arange(0, 11, 1), 'penyampaian')#variabel penyampaian dg nilai 0 samapai 10 dengan range 1,Antecedent adalah library untuk 'Input'
CC = ctrl.Antecedent(np.arange(0, 11, 1), 'citation')#variabel citacion dg nilai 0 samapai 10 dengan range 1,Antecedent adalah library untuk 'Input'
KL = ctrl.Consequent(np.arange(0, 11, 1), 'kelayakan')#variabel kelayakan dg nilai 0 samapai 10 dengan range 1,Consequent adalah library untuk 'Output'
```

# Fungsi Keanggotaan (Membership Function)

Adalah suatu kurva yang menunjukkan pemetaan titik-titik input data kedalam nilai keanggotaan yang memiliki interval antara 0-1.Salah satu cara yang dapat digunakan untuk mendapatkan nilai keanggotaan adalah pendekatan fungsi

# a. fungsi keangggotaan untuk variabel publish (input)

```python
PB['Jarang'] = fuzz.trapmf(PB.universe, [0, 1, 2, 4])#membership function jarang, dalam bentuk trapesium dengan nilai 1 - 4
PB['Sering'] = fuzz.trapmf(PB.universe, [2, 4, 10, 10])#membership function sering, dalam bentuk trapesium dengan nilai 2 - 10
PB.view()#menampilkan gambar
```

# *Berikut adalah persamaan garis pada input :Publish*
1. $PB[Jarang] =
\left\{
    \begin{array}{ll}
       1 & \mbox{; } x \leq 2 \\

       0 & \mbox{; } x \geq 4
    \end{array}
\right.$

2. $PB[Sering] = 
\left\{
    \begin{array}{ll}
       0 & \mbox{; } x \leq 2 \\
       {x-2 \over 4-2} & \mbox{; } 2 \leq x \leq 4 \\

# b. fungsi keangggotaan untuk variabel isi & bahan (input)

```python
IB['Tidak Bagus'] = fuzz.trapmf(IB.universe, [0, 0, 1, 3])#membership function tidak bagus, dalam bentuk trapesium dengan nilai 0 - 3
IB['Cukup'] = fuzz.trimf(IB.universe, [1, 3, 6])#membership function cukup, dalam bentuk segitiga dengan nilai 1 - 6
IB['Bagus'] = fuzz.trapmf(IB.universe, [4, 6, 10, 10])#membership function bagus, dalam bentuk trapesium dengan nilai 4 - 10 
IB.view()#menampilkan gambar
```
