<h1 align = "center"> <font color='green'> A NOISE STRESS TEST FOR ARRHYTMIA DETECTORS</h1> </font>

Noise Stress adalah aspek penting dari kinerja detektor aritmia yang sulit diukur menggunakan teknik konvensional. Dengan menambahkan noise dalam jumlah yang diketahui oleh rekaman EKG beranotasi, kemampuan penanganan noise dari pengukuran kinerja detektor arrythmia bervariasi sebagia fungsi tingkat kebisingan dan jenis kebisingan, dan menjelaskan penggunaannya dalam tes stress kebisingan otomatis dari detektor aritmia

"Noise Stress Test" adalah teknik kuantitatif yang dapat di produksi untuk menilai kinerja detektor aritmia dengan adanya kebisingan dan artefak. Beat by beat evaluasi menggunakan rekaman EKG yang beranotasi, dan diukur dengan kinerja yang konvensional. kebisingan realistis dari jenis yang ditemukan dalam rekaman EKG ditambahkan ke EKG bersih, dan EKG kebisingan yang dihasilkan dianalisis oleh detektor yang sedang diuji. Dengan memvariasikan jenis dan tingkat kebisingan, kinerja detektor dapat diukur sebagai fungsi dari karakteristik kebisingan. Noise Strees Test dirancang agar dapat di produksi secara ketat sehingga efek dari perubahan dalam dektektor dapat dinilai, dan sehingga perbandingan antara detektor dapat dinilai, dan sehingga perbandingan antara dektektor dapat dibuat 

Ide untuk menambahkan kebisingan untuk membersihkan EKG guna menguji toleransi kebisingan dari program analisis EKG diusulkan secara independen oleh dua penulis dan dikembangkan oleh analisis EKG diagnostik Hannover. Kelompok Hannover meneliti efek kebisingan pada keakuratan pengukuran program mereka pada fitur EKG domain waktu. sumber - sumber kebisingan adalah articial, dan termasuk band frekuensi white noise line, low frequency sine dan sawtooth waveforms, serta step function-like artifacts. Sinyal - sinyal EKG dibentuk dengan menyatukan ketukan identik, pada dasarnya bebas dari noise, pada interval R-R yang tetap

### Tipe Model
Mendefinisikan model sequential() yang ditampungkan ke dalam variabel model


```python
model = Sequential()
```

### Input Layer, Hidden Layer, dan  Output Layer
Fungsi yang digunakan pada Input layer adalah ReLU(Rectified Linear Unit(, dimana cara kerjanya yaitu mengubah setiap nilai input negatif menjadi 0 pada grafik dengan sesegera mungkin 
https://machinelearningmastery.com/rectified-linear-activation-function-for-deep-learning-neural-networks/

Fungsi Sigmoid yang sering digunakan pada lapisan akhir dari pengklasifikasian jaringan berbasis saraf. Jaringan ini umumnya ditraining di bawah log loss (*crossentropy*), yang mana memberikan varian nilai non-linear regresi logistik multinomial

```python
model.add(Dense(200, input_dim=252, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(2, activation='sigmoid'))
```

## Compile Model 

Setelah dilakukan proses definisi model, langkah selanjut adalah mengcompile model yang sudah ada tersebut sehingga menjadi satu kesatuan hidden yang biasa disebut dengan *Hidden Layer*. 

***Berikut Contoh gambar dari HIDDEN LAYER***


ketika mengguakan objective function **binary_categorical**, target harus dalam format binary (yakni 0 dan 1, terget untuk setiap sampel harus berupa vektor 2 dimensi)

```python
model.compile(loss='binary_crossentropy', 
              optimizer=adam(lr=0.0001),
              metrics=['accuracy'])
```

## Fit Model 

Sesuai dengan tujuan awal, dataset yang diolah diharapkan mencapai akurasi yang maksimal. Untuk itu diperlukan mengatur parameter - parameter yang diperlukan yakni :

Disini epochs yang dilakukan sebanyak 200 kali

```python
Proses = model.fit(X_scale, Y_categorical, 
                   validation_data=(test_X,test_Y), 
                   epochs=200, batch_size=16)
```

## Visualisasi Proses Training 
 
Dari proses training diatas, setiap loss pada epoch yang terjadi ketika training dapat ditampilkan dalam bentuk grafik. Setiap komponen pada grafik pun dapat ditentukan sesuai keperluan masing masing 

```python
import matplotlib.pyplot as plt
fig,(ax0) = plt.subplots(nrows=1, figsize=(20,8))
ax0.plot(Proses.history['acc'],'red', label='Akurasi Training')
ax0.plot(Proses.history['val_acc'], 'blue', label='Akurasi Validasi')
ax0.plot(label='Accuracy', loc='upper left')
ax0.set_title('Model Accuracy')
ax0.set_xlabel("Epoch")
ax0.set_ylabel("Accuracy")
ax0.legend()
```

## Performance

Setelah mendapatkan hasil singkat dari proses training pada model diatas, selanjutnya dapat dilakukan analisis performance dengan membuat confusion matrix dari Label yang telah diubah menjadi categorical dan nilai prediksi dari model Neural Network terhadap Feature yang telah di **Preprocessing**

```python
CM = confusion_matrix(Y_categorical.argmax(axis=1), predict.argmax(axis=1))
```

```python
CM
```

### 1. Konversi Array CM ke dalam Variabel Float 

Karena data yang didapat dari CM masih dalam bentuk Array, untuk dapat mengkalkulasi nilai tersebut haruslah mengkonversinya kebentuk tipe data integer ataupun float, disini menggunakan variabel yang bertipe float. Disamping kapasitas yang lebih besar dibandingkan integer, tipe data float juga dapat menampilkan nilai berkoma yang nantinya jika akan membuat persentase dari Akurasi, Sensitivity, Specivity, dan Presisi lebih akurat dan tidak ada miss tang dihasilkan 

```python
# True Positif
TP = CM[[0],[0]]
TP = float(TP)

# False Positif
FP = CM[[0],[1]]
FP = float(FP)

# False Negatif
FN = CM[[1],[0]]
FN = float(FN)

# True Negatif
TN = CM[[1],[1]]
TN = float(TN)
```

## Accuary 
Akurasi dalam masalah klasifikasi adalah jumlah prediksi yang tepat yang dibuat oleh model atas semua jenis prediksi yang dibuat.
## Precision
Presisi adalah ukuran yang memberi tahu berapa proporsi pasien yang didiagnosis menderita kanker yang sebenarnya memang menderita Ventrikular atau tidak ada. yang diprediksi mengidap kanker ( True Positif dan False Positif) dan orang - orang yang benar - benar mmengidap Ventrikular (True Positif) 
## Sensitivity
Sensitivity adalah ukuran yang memberitahu beberapa proporsi pasien yang benar - benar menderita Ventrikular didiagnosis oleh Algorythm sebagai Ventrikular. Yang sebenarnya menderiita Ventrikular (True Positif dan False Negatif) dan pasien yang didiagnosis oleh model yang Ventrikular (True Positif).

*Note : False Negatif dimasukkan karena orang itu benar - benar mengidap kanker meskipun modelnya memperkirakan sebaliknya*

## Specificity 
Spesifitas adalah ukuran yang memberitahu bebrapa proporsi pasien yang **TIDAK** menderita **Ventrikular**, diprediksi oleh model sebagai **Non - Ventrikular**. Yang sebenarnya tidak menderita **Ventrikular** (*False Positif dan True Negatif*) dan pasien yang didiagnosis oleh model tidak menderita **Ventrikular** adalah *True Negatif*.

*Note : False Positif disertakan karena pasien **TIDAK** benar - benar menderita **Ventrikular** meskipun modelnya memperkirakan sebaliknya*

```python
print('Accuracy:', ((TP+TN)/(TP+TN+FP+FN)) * 100, '%')
print('Precision:', (TP/(TP+FP) * 100), '%')
print('Sensitivity:', (TP/(TP+FN) * 100), '%')
print('Specificity:', (TN/(TN+FP) * 100), '%')
```
