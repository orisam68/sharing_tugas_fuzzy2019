```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

# Fuzzifikasi

Dalam penentuan jumlah produksi barang berdasarkan data persediaan dan jumlah permintaan, variabel input dibagi menjadi dua yaitu variabel 
1. persediaan 
2. permintaan  
3. Produksi (Output)

```python
Permintaan = ctrl.Antecedent(np.arange(0,2000,10), 'Permintaan')
Persediaan = ctrl.Antecedent(np.arange(0,1500, 10), 'Persediaan')
Produksi = ctrl.Consequent(np.arange(0, 3000, 10), 'Jumlah_produksi')
```

# Fungsi Keanggotaan

Fungsi Keanggotaan ada 3 yaitu : 
1. Sedikit
2. Sedang
3. Banyak

Berikut adalah cara untuk mendapatkan nilai keanggotan berdasarkan variabel linguistik dan variabel numerik yang digunakan: 



\begin{equation}
  ['sedikit'] = \mu sedikit(x)=\begin{cases}

    0, & \text{$x\geq 50$}.
  \end{cases}
  \\
  ['sedang'] = \mu sedang(x)=\begin{cases}
    0, & \text{$x\leq 1030$ atau $x\geq1589$}.\\

  \\
  ['banyak'] = \mu banyak(x)=\begin{cases}

    1, & \text{$x\geq 2000$}.
  \end{cases}
\end{equation}

# Rule Base

Pada tahap ini, nilai keanggotaan himpunan permintaan dan persediaan saat ini dicari menggunakan fungsi keanggotaan himpunan Fuzzy berdasarkan data. Pembentukan Aturan Fuzzy, Dari dua variabel input dan sebuah variabel output yang telah didefinisikan, dengan melakukan analisa data terhadap batas tiap-tiap himpunan Fuzzy pada tiap-tiap variabelnya maka terdapat 9 aturan Fuzzy yang akan dipakai dalam sistem ini.

|Rule|Permintaan|Persediaan|Produksi|
|----|----------|----------|--------|
|Rule1|Sedikit|Sedikit|Sedikit|
|Rule2|Sedikit|Sedang|Sedang|
|Rule3|Sedikit|Banyak|Banyak|
|Rule4|Sedang|Sedikit|Sedang|
|Rule5|Sedang|Sedang|Sedang|
|Rule6|Sedang|Banyak|Banyak|
|Rule7|Banyak|Sedikit|Banyak|
|Rule8|Banyak|Sedang|Banyak|
|Rule9|Banyak|Banyak|Banyak|

```python
rule1 = ctrl.Rule(Permintaan['sedikit'] & Persediaan['sedikit'], Produksi['sedikit'])
rule2 = ctrl.Rule(Permintaan['sedikit'] & Persediaan['sedang'], Produksi['sedang'])
rule3 = ctrl.Rule(Permintaan['sedikit'] & Persediaan['banyak'], Produksi['banyak'])

rule4 = ctrl.Rule(Permintaan['sedang'] & Persediaan['sedikit'], Produksi['sedang'])
rule5 = ctrl.Rule(Permintaan['sedang'] & Persediaan['sedang'], Produksi['sedang'])
rule6 = ctrl.Rule(Permintaan['sedang'] & Persediaan['banyak'], Produksi['banyak'])

rule7 = ctrl.Rule(Permintaan['banyak'] & Persediaan['sedikit'], Produksi['banyak'])
rule8 = ctrl.Rule(Permintaan['banyak'] & Persediaan['sedang'], Produksi['banyak'])
rule9 = ctrl.Rule(Permintaan['banyak'] & Persediaan['banyak'], Produksi['banyak'])
```

```python
nilai_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
```

# Defuzzifikasi

Nilai inputan permintaan yaitu 1308 dan
Nilai inputan Persediaan yaitu 700

```python
Nilai = ctrl.ControlSystemSimulation(nilai_ctrl)

Nilai.input['Permintaan'] = 1308
Nilai.input['Persediaan'] = 700

Nilai.compute()

print(Nilai.output['Jumlah_produksi'])

Produksi.view(sim=Nilai)

plt.show()
```
