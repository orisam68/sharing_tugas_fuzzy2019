#### 5. Penyelesaian Case Fuzzy
Pada penyelesaian case fuzzy ini, import library yang digunakan yaitu :

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

Memberikan Arange Variabel Input

```python
permukaan_lantai = ctrl.Antecedent(np.arange(0,200,1),'permukaan lantai')
jumlah_debu = ctrl.Antecedent(np.arange(0,200,1),'jumlah debu')
```

Memberikan Arange Variabel Output

```python
power = ctrl.Consequent(np.arange(0,30,1),'power')
```

Menentukan Membership Function dengan menggunakan Triangular Membership Function

1. Membership Function Input : 

- Membership Function dari Permukaan Lantai

```python
permukaan_lantai['halus'] = fuzz.trapmf(permukaan_lantai.universe,[0,0,30,80])
permukaan_lantai['sedang'] = fuzz.trimf(permukaan_lantai.universe,[70,110,150])
permukaan_lantai['kasar'] = fuzz.trapmf(permukaan_lantai.universe,[120,180,200,200])
```

- Menampilkan Hasil Plotting dari Tahap Membership Function Permukaan Lantai

```python
permukaan_lantai.view()
```

- Membership Function Jumlah Debu

```python
jumlah_debu['sedikit'] = fuzz.trapmf(jumlah_debu.universe,[0,0,20,90])
jumlah_debu['sedang'] = fuzz.trimf(jumlah_debu.universe,[80,120,160])
jumlah_debu['banyak'] = fuzz.trapmf(jumlah_debu.universe,[130,190,200,200])
```

- Menampilkan Hasil Plotting dari Tahap Membership Function Jumlah Debu

```python
jumlah_debu.view()
```

2. Membership Function Output

- Membership Function dari Power

```python
power['lemah'] = fuzz.trapmf(power.universe,[0,0,5,12])
power['sedang'] = fuzz.trimf(power.universe,[8,13,18])
power['kuat'] = fuzz.trapmf(power.universe,[15,25,30,30])
```

- Menampilkan Hasil Plotting dari Tahap Membership Function Power

```python
power.view()
```

#### 5. Inferencing 

Inferencing merupakan suatu proses menentukan rule dari kasus fuzzy yang kita buat dengan kondisi IFThen.
Pada tahap inferencing ini, saya membuat 9 rule dikarenakan fuzzifikasi permukaan lantai berjumlah 3 dan fuzzifikasi jumlah debu yang ada berjumlah 3, maka 3x3 = 9 rule base.


#### Rule Base
Adapun aturan rule base yang digunakan , sebagai berikut :



```python
rule1 = ctrl.Rule(jumlah_debu['banyak'] & permukaan_lantai['kasar'],power['kuat'])
rule2 = ctrl.Rule(jumlah_debu['sedang'] & permukaan_lantai['kasar'],power['kuat'])
rule3 = ctrl.Rule(jumlah_debu['sedikit'] & permukaan_lantai['kasar'],power['sedang'])
rule4 = ctrl.Rule(jumlah_debu['banyak'] & permukaan_lantai['sedang'],power['kuat'])
rule5 = ctrl.Rule(jumlah_debu['sedang'] & permukaan_lantai['sedang'],power['sedang'])
rule6 = ctrl.Rule(jumlah_debu['sedikit'] & permukaan_lantai['sedang'],power['lemah'])
rule7 = ctrl.Rule(jumlah_debu['banyak'] & permukaan_lantai['halus'],power['sedang'])
rule8 = ctrl.Rule(jumlah_debu['sedang'] & permukaan_lantai['halus'],power['lemah'])
rule9 = ctrl.Rule(jumlah_debu['sedikit'] & permukaan_lantai['halus'],power['lemah'])
```

```python
rule1.view()
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
rule7.view()
rule8.view()
rule9.view()
```

#### Rule Evaluation

Pada tahap rule evaluation dilakukan evaluasi, pengecekan, pengambilan keputusan aturan, knowledge base, rule base yang akan diterapkan dengan menyesuaikan kondisi nilai pada himpunan-himpunan fuzzy.

```python
power_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9])
```

#### 6. Defuzzifikasi

```python
powerr = ctrl.ControlSystemSimulation(power_ctrl)
```

Memasukkan nilai input dari permukaan lantai dan input jumlah debu

```python
powerr.input['permukaan lantai']=100
```

```python
powerr.input['jumlah debu'] = 100
```

Perhitungan dari defuzzifikasi

```python
powerr.compute()
```

Menampilkan hasil output defuzzifikasi

```python
print(powerr.output['power'])
```

```python
permukaan_lantai.view(sim=powerr)
```

```python
jumlah_debu.view(sim=powerr)
```

```python
power.view(sim=powerr)
```
